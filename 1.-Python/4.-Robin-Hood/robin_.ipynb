```python
points = [(4, 5), (-0, 2), (4, 7), (1, -3), (3, -2), (4, 5), (3, 2), (5, 7), (-5, 7), (2, 2), (-4, 5), (0, -2),
          (-4, 7), (-1, 3), (-3, 2), (-4, -5), (-3, 2), (5, 7), (5, 7), (2, 2), (9, 9), (-8, -9)]

dups = set(points)
#print(len(dups),len(points))
print("1. coordinates where robin hood hit more than once: \n",dups)
dups=list(dups)
#---------------------
print("\n2. arrows per quadrant:")
x_pos=[]
x_neg=[]
Xpos_Ypos=[]
Xpos_Yneg=[]
Xneg_Ypos=[]
Xneg_Yneg=[]
#-divide main list by X positive or negative
for a in range(len(dups)):
    if dups[a][0] >=0:
        x_pos.append(dups[a])
    else:
        x_neg.append(dups[a])

#-divide X_positive by Y pos or neg
for b in range(len(x_pos)):
    if x_pos[b][1] >=0:
        Xpos_Ypos.append(x_pos[b])
    else:
        Xpos_Yneg.append(x_pos[b])

#-divide X_negative by Y pos or neg
for c in range(len(x_neg)):
    if x_neg[c][1] >=0:
        Xneg_Ypos.append(x_neg[c])
    else:
        Xneg_Yneg.append(x_neg[c])

print("\n\tXpos_Ypos len:",len(Xpos_Ypos))
print("\n\tXpos_Yneg len:",len(Xpos_Yneg))
print("\n\tXneg_Ypos len:",len(Xneg_Ypos))
print("\n\tXneg_Yneg len:",len(Xneg_Yneg))
#---------------------
print("\n3. point closest to the center is: ")
sqrt=[]
no_list=[]
for i in range(len(dups)):
    zeta = ((dups[i][0])**2 + (dups[i][1])**2)**0.5
    sqrt.append(zeta)
chiqui = min(sqrt)

for n in range(len(sqrt)):
    if sqrt[n] == chiqui:
        no_list.append(n)
    n+=1

for mm in range(len(no_list)):
    print(dups[no_list[mm]])
#---------------------
#print("\n4. arrows that won't hit target:")
no_hit=[]
for p in range(len(points)):
    peta = ((points[p][0])**2 + (points[p][1])**2)**0.5
    if peta > 3.0:
        no_hit.append(peta)
        #print(points[p],"",peta)

print("\n4. arrows that won't hit target:\n",len(no_hit))
```
